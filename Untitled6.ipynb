```python

def preprocessImage(image):
  #Gray scale the image
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  # filter the image
  
  image_mean = gray_image.mean()
  image_std = gray_image.std()
  indeces = gray_image < image_mean + image_std / 2
  gray_image[indeces] = 0
  #normalize the image
  width ,height = gray_image.shape
  widht = width * 2
  height = height * 2
  new_image = cv2.resize(gray_image,(width, height), interpolation= cv2.INTER_LINEAR)
  norm = np.zeros((600,600))
  final_image = cv2.normalize(new_image,  norm, 0, 255, cv2.NORM_MINMAX)
  return final_image

def thick_format(image):
  import numpy as np
  kernel = np.ones((2,2), np.uint8)
  image = cv2.dilate(image,kernel, iterations = 1)
  image = cv2.bitwise_not(image)
  return image

def thin_format(image):
  import numpy as np
  kernel = np.ones((1,1), np.uint8)
  image = cv2.erode(image,kernel, iterations = 1)
  image = cv2.bitwise_not(image)

  return image


def returnTextFromImage(processedimage):
  words = []
  results = reader.readtext(processedimage)

  for blocks in results:
    words.append(blocks[1])

  return words
```

```python
reader = easyocr.Reader(lang_list=['en'])
corpus = []
for path in file_paths:
  im = cv2.imread(path)
  processed_image = preprocessImage(im)

  words = returnTextFromImage(processed_image)
  for word in words:
    corpus.append(word)



```
